# GDP_per_capita

```python
import pandas as pd
file = pd.read_csv('List of Countries by GDP (nominal) per Capita.csv')
file = file.drop('Unnamed: 0', axis=1)
file = file.drop(0, axis=0)
file
```

```python
country=file.iloc[:,0]
region=file.iloc[:,1]

imf_estimate=file.iloc[:,2]
imf_year=file.iloc[:,3]
df_imf = pd.DataFrame(list(zip(country,region,imf_estimate,imf_year)),columns=['Country/Territory','UN Region','Estimate','Year'])
print('List of Countries by GDP (nominal) per Capita by IMF')
df_imf.sort_values(by=['UN Region'])
```

```python
worldbank_estimate=file.iloc[:,4]
worldbank_year=file.iloc[:,5]
df_worldbank = pd.DataFrame(list(zip(country,region,worldbank_estimate,worldbank_year)),columns=['Country/Territory','UN Region','Estimate','Year'])
print('List of Countries by GDP (nominal) per Capita by World Bank')
df_worldbank
```

```python
unitednations_estimate=file.iloc[:,6]
unitednations_year=file.iloc[:,7]
df_unitednations = pd.DataFrame(list(zip(country,region,unitednations_estimate,unitednations_year)),columns=['Country/Territory','UN Region','Estimate','Year'])
print('List of Countries by GDP (nominal) per Capita by United Nations')
df_unitednations
```
